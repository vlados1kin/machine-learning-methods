```python
def nod(a,b):
  while b:
    a, b = b, a % b
  return a
a = 24
b = 36
res = nod(a, b)
print(res)
```

```python
def type_checker(data):
  if isinstance(data, dict):
    sorted_items = sorted(data.items(), key=lambda item: item[1])
    top_three_items = sorted_items[:3]
    top_three_keys = [item[0] for item in top_three_items]
    return top_three_keys
  elif isinstance(data, list):
    while 0 in data:
      data.remove(0)
    mux = 1;
    counter = 0;
    for el in data:
      if el > 0 and counter < 2:
        mux *= el
        counter += 1
    print(data)
    return mux
  elif isinstance(data, int):
    return len(str(data))
  elif isinstance(data, str):
    char_counter = {}
    for char in data:
      if char in char_counter:
        char_counter[char] += 1
      else:
        char_counter[char] = 1
    return char_counter

my_dict = {"key_1": 500, 2: 400, "key_3": 100, 4: 250}
print(type_checker(my_dict))
my_list = [0,-1,7,0,2,0,0,0,3]
print(type_checker(my_list))
my_number = 1324353328
print(type_checker(my_number))
my_string = "aaaa1111bvfbvfv"
print(type_checker(my_string))
```
