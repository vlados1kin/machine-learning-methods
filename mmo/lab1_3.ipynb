```python
class Calculator:
  def sum(self, a, b):
    return a + b
  def sub(self, a, b):
    return a - b
  def mux(self, a, b):
    return a * b
  def div(self, a, b):
    return a / b

  def calculate(self):
    a = int(input("Введите первое число: "))
    b = int(input("Введите первое число: "))
    operator = input("Введите знак операции: ")
    if operator == "+":
      print(f"{a} + {b} = {self.sum(a, b)}")
    elif operator == "-":
      print(f"{a} + {b} = {self.sub(a, b)}")
    elif operator == "*":
      print(f"{a} + {b} = {self.mux(a, b)}")
    elif operator == "/":
      print(f"{a} + {b} = {self.div(a, b)}")
    else:
      print("Неверный ввод.")

calc = Calculator()
while True:
  calc.calculate()
  next = input("Хотите продолжить(да/нет): ")
  if next != "да":
    break
```

```python
class Car:
  def __init__(self, speed, color, name, is_police):
    self.speed = speed
    self.color = color
    self.name = name
    self.is_police = is_police
  def go(self, new_speed):
    self.speed = new_speed
    print(f"{self.name} поехала")
  def stop(self):
    self.speed = 0
    print(f"{self.name} остановилась")
  def turn(self, direction):
    print(f"{self.name} повернула {direction}")
  def show_speed(self):
    print(f"{self.name} едет со скоростью {self.speed}")

class TownCar(Car):
  def __init__(self, speed):
    super().__init__(speed, "черный", "TownCar", False)
  def show_speed(self):
    if self.speed > 60:
      print(f"{self.name} превышает скорость ({self.speed})")
    else:
      print(f"{self.name} едет со скоростью {self.speed}")

class SportCar(Car):
  def __init__(self, speed):
    super().__init__(speed, "красный","SportCar", False)

class WorkCar(Car):
  def __init__(self, speed):
    super().__init__(speed, "серый","WorkCar", False)

  def show_speed(self):
    if self.speed > 40:
      print(f"{self.name} превышает скорость ({self.speed})")
    else:
      print(f"{self.name} едет со скоростью {self.speed}")

class PoliceCar(Car):
  def __init__(self, speed):
    super().__init__(speed, "синий","PoliceCar", True)

car1 = TownCar(80)
car1.show_speed()
car1.stop()
car1.show_speed()
car1.go(40)
car1.show_speed()


car2 = WorkCar(40)
car2.show_speed()
car2.turn("налево")
car2.stop()

```

```python
import string
class Alphabet:
  def __init__(self, lang, letters):
    self.lang = lang
    self.letters = letters
  def print(self):
    print(self.letters)
  def letters_num(self):
    return len(self.letters)

class EngAlphabet(Alphabet):
  def __init__(self):
    super().__init__("En", string.ascii_uppercase)
    self.__letters_num = len(self.letters)
  def is_en_letter(self, letter):
    return letter in self.letters
  def letters_num(self):
    return self.__letters_num

  @staticmethod
  def example():
    return "Hello world"

alph = EngAlphabet()
alph.print()
print(alph.is_en_letter("S"))
print(alph.letters_num())
print(EngAlphabet.example())
```
