```python
user_input = input("Введите строку ")
arr = user_input.split()
a = 0
b = 0
for el in arr:
    if len(el) % 2 == 0:
      a += 1
    else:
      b += 1
print("Кол-во слов четной длины:", a)
print("Кол-во слов нечетной длины:", b)
```

```python
user_input = input("Введите строку ")
dict = {}
for char in user_input:
  if dict.get(char) == None:
    dict.update({char: user_input.count(char)})
print(dict)
```

```python
menu = {"Торт":[["Сахар","Мука","Яйца","Корица","Крем"], 20, 200],
        "Зефир":[["Сахар","Яблоко"], 10, 100],
        "Маффин":[["Сахар","Мука","Яйца","Корица"],15,500]}
while True:
  print("1. Просмотр описания.")
  print("2. Просмотр цены.")
  print("3. Просмотр количества.")
  print("4. Вся информация.")
  print("5. Покупка.")
  print("6. До свидания")
  command = int(input())
  if command == 1:
    for key, value in menu.items():
      print(key, " - ", value[0])
  elif command == 2:
    for key, value in menu.items():
      print(key, " - ", value[1], " за 100 гр")
  elif command == 3:
    for key, value in menu.items():
      print(key, " - ", value[2], " гр")
  elif command == 4:
    for key, value in menu.items():
      print(key, " - ", value[0],", ",  value[1], " за 100 гр,", value[2], " гр")
  elif command == 5:
    item = input("Введите название товара ")
    number = int(input("Введите количество товара "))
    if menu.get(item) != None and menu.get(item)[2] >= number:
      menu.get(item)[2] -= number
      print("Цена выбранных товаров: ", menu.get(item)[1]*number)
      print(menu.get(item))
    else:
      print("Неверный ввод")
  else:
    break
```

```python
from random import randint

attempts = 0
number = randint(0,100)
while True:
    attempts += 1
    current_number = int(input("Введите число от 0 до 100 "))
    if current_number == number:
      break
    elif current_number > number:
      print("Ваше число больше загаданного")
    elif current_number < number:
      print("Ваше число меньше загаданного")
print("Вы угадали с %s попытки" % attempts)
```
